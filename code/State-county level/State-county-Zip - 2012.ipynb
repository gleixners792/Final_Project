```python
# Dependencies

import pandas as pd
import numpy as np
```

```python
# Read data file and store into pandas data frame.

df = pd.read_csv("2012/Medicare_Provider_Util_Payment_PUF_CY2012_update/Medicare_Provider_Util_Payment_PUF_CY2012.txt", sep="\t", header=0)
```

```python
# Check data

df.dtypes
```

```python
# Check Data

len(df)
```

```python
# Select only US

df1= df[df.NPPES_PROVIDER_COUNTRY == 'US']
df1
```

```python
# Create ZIP Column

df1['ZIP'] = df1['NPPES_PROVIDER_ZIP']
df1.head(5)
```

```python
# Check Data

df1.dtypes
```

```python
# Check unique ZIPs

df1['ZIP'].unique
```

```python
# Create new dataframe with selected columns

df2 = df1[['ZIP','NPPES_PROVIDER_STATE', 'NPI', 'PLACE_OF_SERVICE','PROVIDER_TYPE', 'HCPCS_CODE', 'HCPCS_DRUG_INDICATOR', 'AVERAGE_SUBMITTED_CHRG_AMT', 'STDEV_MEDICARE_PAYMENT_AMT']]
df2.tail(50)
```

```python
# Check New dataframe

df2.dtypes
```

```python
# Create states list to then exclude US territories

states = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DC", "DE", "FL", "GA", 
          "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", 
          "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", 
          "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", 
          "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]
df2 = df2[df2['NPPES_PROVIDER_STATE'].isin(states)]
df2
```

```python
# Clean and sort ZIP codes for merging

df2['ZIP'] = df2['ZIP'].astype(str).str[0:5].map(lambda x: str(x.replace('.','')))
df2['ZIP'] = df2['ZIP'].astype(int)
df2 = df2.dropna()
df2.sort_values(by=['ZIP'], inplace=True)
df2
```

```python
# Check data

len(df2)
```

```python
# Check data

df2.dtypes
```

```python
# Read ZIP data file and store into pandas data frame

df3 = pd.read_csv('../Final_project_BigData/ZIP-COUNTY-FIPS.csv') 
df3.dtypes
```

```python
# Check data

len(df3)
```

```python
# Create states list to then exclude US territories

df3 = df3[df3['STATE'].isin(states)]
df3
```

```python
# Select column for the merge

df3 = df3[['ZIP','COUNTYNAME']]
df3.sort_values(by=['ZIP'], inplace=True)
df3
```

```python
# Check data

df3.dtypes
```

```python
# Merge dataframes by Zipcode

dfinal= pd.merge_asof(df2, df3, on='ZIP')
```

```python
# Sort Data by NPI number

dfinal.sort_values(by=['NPI'], inplace=True)
dfinal.head(100)
```

```python
# Check data

len(dfinal)
```

```python
# Create Dataframe 

df_2012 = dfinal.groupby(['NPPES_PROVIDER_STATE', 'COUNTYNAME', 'ZIP', 'PROVIDER_TYPE', 'PLACE_OF_SERVICE']).agg({'AVERAGE_SUBMITTED_CHRG_AMT': ['mean'], 'STDEV_MEDICARE_PAYMENT_AMT': ['mean']})
df_2012.columns = ['avg_submit_2012', 'avg_payment_2012']
df_2012 = df_2012.reset_index()

df_2012.head(50)
```

```python
# Check data

len(df_2012)
```

```python
# Save dataframe as CSV

df_2012.to_csv('2012.csv') 
```
