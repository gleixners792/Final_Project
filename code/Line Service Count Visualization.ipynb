```python
import os
import pandas as pd
import numpy as np
```

# Read the CSV and Perform Basic Data Cleaning

```python
input_data_file = "output_data_npi.txt"
df = pd.read_csv(os.path.join("..", "Data", input_data_file),sep='\t')
#df = pd.read_csv(os.path.join("..", "Data", input_data_file),sep='\t', nrows=500000)

# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
#df = df.dropna()
df.head()
```

```python
df.info()
```

```python
df.drop(df[df['npi'] == 1].index, inplace = True)
df.drop(df[df['nppes_provider_country'] != "US"].index, inplace = True)

df.drop(df[df['nppes_provider_state'] == "AA"].index, inplace = True)
df.drop(df[df['nppes_provider_state'] == "AE"].index, inplace = True)
df.drop(df[df['nppes_provider_state'] == "AP"].index, inplace = True)
df.drop(df[df['nppes_provider_state'] == "AS"].index, inplace = True)
df.drop(df[df['nppes_provider_state'] == "DC"].index, inplace = True)
df.drop(df[df['nppes_provider_state'] == "GU"].index, inplace = True)
df.drop(df[df['nppes_provider_state'] == "MP"].index, inplace = True)
df.drop(df[df['nppes_provider_state'] == "PR"].index, inplace = True)
df.drop(df[df['nppes_provider_state'] == "VI"].index, inplace = True)
df.drop(df[df['nppes_provider_state'] == "XX"].index, inplace = True)
```

```python
new_df = pd.pivot_table(df, values= "line_srvc_cnt", index= "nppes_provider_state", columns="year", aggfunc=np.sum, fill_value=0, margins=True)
```

```python
new_df
```

```python
new_df = new_df.sort_values("All", ascending=False)
new_df = new_df.drop(['All'], axis=1)
new_df = new_df.drop(index='All')
new_df
```
