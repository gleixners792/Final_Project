```python
#data_df = pd.read_csv(os.path.join("..", "Data", data_file2012),sep='\t', nrows=50000)
data_df = pd.read_csv(os.path.join("..", "Data", data_file2012),sep='\t')
# Drop the null rows
#data_df = data_df.dropna()
#data_df.head()
data_df.info()
```

```python
data_df.drop(data_df[data_df['NPI'] == 1].index, inplace = True)
data_df.drop(data_df[data_df['NPPES_PROVIDER_COUNTRY'] != "US"].index, inplace = True)
data_df.info()
```

```python
data_df['NPPES_PROVIDER_ZIP'] = data_df['NPPES_PROVIDER_ZIP'].apply(str)
#data_df.info()
data_df['new_zip'] = data_df['NPPES_PROVIDER_ZIP'].str[:5]
df_data_2012 = data_df.groupby(['new_zip', 'PROVIDER_TYPE']).agg({'AVERAGE_SUBMITTED_CHRG_AMT': ['mean'], 'AVERAGE_MEDICARE_PAYMENT_AMT': ['mean']})
df_data_2012.columns = ['avg_submit_2012', 'avg_payment_2012']
df_data_2012 = df_data_2012.reset_index()

#df_data_2012
```
